# Pandas Library

Pandas is an open-source, BSD-licensed Python library providing high-performance, easy-to-use data structures and data analysis tools for the Python programming language. Python with Pandas is used in a wide range of fields including academic and commercial domains including finance, economics, Statistics, analytics, etc.

The name Pandas is derived from the word Panel Data – an Econometrics from Multidimensional data.

```python
#import pandas library
import pandas as pd
```

Pandas deals with the following  data structures −

+ Series 
+ DataFrame 

All Pandas data structures are value mutable (can be changed) and except Series all are size mutable. Series is size immutable.

## Series 

Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.). The axis labels are collectively called index. 

```python
data = [1, 2, 3, 4 ]
series = pd.Series(data)
print(series)
```

```python
type(series)
```

```python
# Setting the index of a series
series1 = pd.Series(data, index=['a','b','c','d'])
print(series1)
```

## DataFrame

A DataFrame is a 2-dimensional data structure that can store data of different types (including characters, integers, floating point values, categorical data and more) in columns. It is similar to a spreadsheet, a SQL table or the data.frame in R. Each column in a DataFrame is a Series. 

```python
#creating dataframe using list
df = pd.DataFrame(data)
print(df)
```

```python
#creating dataframe using dictionary
dict = {'fruits':['apples','banana','mango'],'count':[10,20,15]}
df = pd.DataFrame(dict)
print(df)
```

```python
#creating dataframe using series
df = pd.DataFrame(series1)
print(df)
```

### Merge

Merge DataFrame objects by performing a database-style join operation by columns or indexes.

If joining columns on columns, the DataFrame indexes will be ignored. Otherwise if joining indexes on indexes or indexes on a column or columns, the index will be passed on.

```python
player = ['Player1','Player2','Player3']
point = [8,9,7]
title = ['Game1','Game2','Game3']
df1 = pd.DataFrame({'Player':player,'Points' : point, 'Title' : title },index =['L1','L2','L3'])
print(df1)
```

```python
player = ['Player1','Player5','Player6']
power = ['Punch','Kick','Elbow']
title = ['Game1','Game5','Game6']
df2 = pd.DataFrame({'Player':player,'Power' : power, 'Title' : title },index =['L2','L3','L4'])
print(df2)
```

```python
player = ['Player1','Player5','Player6']
power = ['Punch','Kick','Elbow']
title = ['Game1','Game5','Game6']
df3 = pd.DataFrame({'Players':player,'Power' : power, 'Titles' : title },index =['L2','L3','L4'])
print(df3)
```

```python
type(df1)
```

```python
# Inner Merge
df1.merge(df2, on='Player', how='inner')
```

```python
# Left Merge
df1.merge(df2, on='Player', how='left')
```

```python
# Right Merge
df1.merge(df2, on='Player', how='right')
```

```python
# Outer Merge
df1.merge(df2, on='Player', how='outer')
```

### Join

Join items with other Panel either on major and minor axes column. Joined is based on index. Common index needed between the dfs 

```python
# Inner Join
df1.join(df3, how='inner')
```

```python
# Right Join
df1.join(df3, how='right')
```

```python
# Left Join
df1.join(df3, how='left')
```

```python
# Outer Join
df1.join(df3, how='outer')
```

### Concatenate

Concatenate pandas objects along a particular axis with optional set logic along the other axes.

Can also add a layer of hierarchical indexing on the concatenation axis, which may be useful if the labels are the same (or overlapping) on the passed axis number.

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df3])
```

# Cleaning 

Data cleaning or data cleansing is the process of detecting, correcting or removing corrupt or inaccurate records from a data-set, table or database. It also refers to identifying incomplete, incorrect, inaccurate or irrelevant parts of the data and then replacing, modifying or deleting the dirty data.

After analysing we have realized that our dataset is not perfect, some column names are irrelevant, there are a few missing data, some column has important data but cant perform some operation because they are of string type. So now with the help of Pandas we will be cleaning or in other words we will be making our data perfect to perform further operations.

#### Rename unnamed column

```python
cars=cars.rename(columns={'Unnamed: 1':'model'})
cars
```

#### Replace null value with mean 

```python
#Fill the null values with mean of the column
cars.qsec=cars.qsec.fillna(cars.qsec.mean())
cars
```

#### Drop unwanted column

```python
cars = cars.drop(columns=['S.No'])
cars
```

#### Find variable correlation 

Correlation matrix will help to understand how one variable is related to another. Based on correction value we can decide the variables to be included in the analysis. 

```python
#find correlation matrix
df=cars[['mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']].corr()
df
#you can see mpg is of string type so we can't perform correlation.
```

#### Variable type conversion

```python
#So change mpg from string to float to perform correlation
cars.mpg = cars.mpg.astype(float)
#see the change
cars.info(null_counts=True)
```

```python
#Again find the correlation including mpg
df=cars[['mpg', 'cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']].corr()
df
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(15, 10))
sn.heatmap(df, annot=True)
plt.show()
```

Now from the correlation we can see lowest correlation is between "am" and "carb". So when we will taking "carb" as reference for any analysis we can drop "am", since it will not add any change or value to the analysis. In the same way we can drop "carb" column while taking "am" as reference for any analysis.
