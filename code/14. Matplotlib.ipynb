```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt

# Instead of using "fig = plt.figure(figsize = (18,6))" to set the figure size for different plots, you can use the below line 
# to set the figure size for the whole set of plots along with the import statements.
plt.rcParams["figure.figsize"] = (18,6)
```

## Line Plot



```python
# defining the dataset
x = np.arange(0,10,1)
y = 2 * x + 5

# plotting the datapoints
plt.plot(x,y)
plt.show()
```

### Cutomizing the Line Plot

```python
# changing the figure

# plotting the datapoints
plt.plot(x,y,linewidth = 2.0,linestyle = ":", color = 'r',alpha = 0.5,marker ='o')

# Adding title, label and legend
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1'],loc='best')

# Adding grid
plt.grid(True)
plt.show()
```

### Create two or more plots in a fig

```python
y1 = 2 * x + 5
y2 = 3 * x + 10

# Create two or more plots in a fig
plt.subplot(1,2,1)
# PLotting the points
plt.plot(x,y1)
plt.title('Graph1')

# PLotting the points
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Graph2')

plt.show()
```

```python
# Create two or more plots in a fig
plt.subplot(2,1,1)
# PLotting the points
plt.plot(x,y1)
plt.title('Graph1')

# PLotting the points
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.title('Graph2')

plt.show()
```

```python
# Create two plots in a fig
plt.plot(x,y1, color = 'r')
plt.plot(x,y2, color = 'g')
plt.title('Combined Graph')

plt.show()
```

## Bar Plot

```python
#creating the dataset
data = {'apples':20,'Mangoes':15,'Lemon':30,'Oranges':10}
names = list(data.keys())
values = list(data.values())

#Creating the bar plot
fig = plt.figure(figsize = (10,5))
plt.bar(names,values)
plt.show()
```

### Customizing the Bar Chart

```python
#Creating the bar plot
fig = plt.figure(figsize = (10,5))
plt.barh(names,values,color="orange")
plt.title("Bar Graph Demo")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()
```

## Scatter Plot

```python
# Creating the dataset
a =[10,20,30,40,50,60,70,80]
b =[5,3,2,5,6,1,4,2]
x =[1,2,3,4,3,2,1,2]

# Creating the scatter Plot
plt.scatter(a,b)
plt.scatter(a,x)
plt.show()
```

### Customizing our Scatter Plot

```python
# Creating the scatter Plot
plt.scatter(a,b, c='g', s=100,edgecolors='y',marker='o',alpha =0.5)
plt.scatter(a,x,c='r', s=400,edgecolors='b',marker='4',alpha =1)

plt.legend(['b','x'],loc='best')
plt.title("Scatter Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.grid(True)

# Save the scatter plot as a file
plt.savefig('Scatter Plot.png') 
plt.show()
```

## Histogram

```python
# prepraring the dataset

numbers = [10,12,16,19,11,20,26,28,30,38,35,34,45,60,68,64,62,70,78,75,79,85,94,95]
# creating the histogram
plt.hist(numbers, bins = [0,20,40,60,80,100])
plt.show()
```

### Customizing the Histogram

```python
# creating the histogram
plt.hist(numbers, bins = [0,20,40,60,80,100], edgecolor='#000000', color ='#FF2331')
plt.title("Histogram Demo")
plt.xlabel("Range of values")
plt.ylabel("Frequency of values")
plt.grid(True)
plt.savefig("Histogram Plot.png")
plt.show()
```

## Box Plot

```python
# Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])

# Plotting the data
plt.boxplot(data, showmeans = True)
plt.title("Box Plot Demo")
plt.grid(True)
plt.show()
```

## Violin plot

```python
# Plotting the data
plt.violinplot(data, showmeans = True, showmedians = True)
plt.title("Violin Plot Demo")
plt.grid(True)
plt.show()
```

## Pie Chart

```python
# Preparing the dataset
label = ['Dog','Cat','Wolf','Lion']
sizes = [50,45,60,80]

# Creating the pie chart 
plt.pie(sizes, labels = label,autopct='%1.1f%%',shadow=True, startangle = 90,explode = (0,0.1,0,0))
plt.title("Pie Chart Demo")
plt.show()
```

## Doughnut Chart

```python
size_centre = [5]
a,b,c,d = [plt.cm.magma,plt.cm.Reds,plt.cm.coolwarm,plt.cm.Spectral]
#Creating the doughnut chart
pie1 = plt.pie(sizes,labels = label, radius = 1.5,colors = [a(0.5),b(1.5),c(1.0),d(2.0)])
pie2 = plt.pie(size_centre,radius = 1.0,colors = 'w')
plt.show()
```

## Area Chart

```python
#define the dataset
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]

# Creating the area plot
plt.stackplot(x,y,colors = 'g', alpha = 0.25)
# Merge area chart with the line chart
plt.plot(x,y)
plt.grid(True)
plt.show()
```
